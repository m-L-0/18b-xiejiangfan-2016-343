```python
from numpy import *
import operator
from sklearn import datasets
from sklearn.model_selection import train_test_split

#导入数据
def LoadDataSet():
    iris = datasets.load_iris()
    X = iris.data
    y = iris.target
    return X,y

#定义KNNClassfier类
class KNNClassfier():
    def __init__(self, data, labels):
        self.data = data
        self.label = labels
    #归一化特征值
    def AutoNorm(self):
        dataSet = self.data
        minVals = dataSet.min(0)
        maxVals = dataSet.max(0)
        ranges = maxVals - minVals
        normDataSet = zeros(shape(dataSet))
        m = dataSet.shape[0]
        normDataSet = dataSet - tile(minVals, (m,1))
        normDataSet = normDataSet/tile(ranges, (m,1))
        return normDataSet

    #knn计算
    def KnnClassify(self, inX,dataSet,labels, k):
        dataSetSize = dataSet.shape[0]
        #距离计算
        diffMat = tile(inX, (dataSetSize,1)) - dataSet
        sqDiffMat = diffMat**2
        sqDist = sqDiffMat.sum(axis=1)
        dist = sqDist**0.5
        sortedDistIndex = dist.argsort()
        classCount = {}
        #选择距离最小的k个点
        for i in range(k):
            voteLabel = labels[sortedDistIndex[i]]
            classCount[voteLabel] = classCount.get(voteLabel, 0) + 1

        sortedClassCount = sorted(classCount.items(), key = operator.itemgetter(1), reverse = True)
        print(sortedClassCount)
        return sortedClassCount[0][0]

    #测试
    def DataClassTest(self, X_train, X_test,y_trian,y_test):
        m = X_test.shape[0]
        errorCount = 0.0
        for i in range(m):
            predict_y  = self.KnnClassify(X_test[i,:], X_train, y_train, 3)
            print("the classifier came back with: %d, the real answer is: %d"\
                 %(predict_y, y_test[i]))
            if predict_y != y_test[i]:
                errorCount += 1.0
            print("the total error rate is:%f" %(errorCount/float(m))) #错误率
if __name__ == '__main__':
    X, y = LoadDataSet()   #导入数据
    K = KNNClassfier(X, y) #构造类
    K.data = K.AutoNorm()  #标准化预处理
    X_train, X_test, y_train, y_test = train_test_split(K.data, K.label, test_size=0.2, random_state=7,stratify=y) #划分训练集和测试集
    K.DataClassTest(X_train, X_test, y_train, y_test)  #进行测试，输出错误率
```

```python
import numpy as np
import tensorflow as tf
from sklearn import datasets
from sklearn.model_selection import train_test_split

#导入鸢尾花数据集
iris = datasets.load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7,stratify=y)

#图的输入
train2_X = tf.placeholder("float",[None,4])
test2_X = tf.placeholder("float",[4])

#L1求距离
distance = tf.reduce_sum(tf.abs(tf.add(train2_X, tf.negative(test2_X))),reduction_indices=1)
#reduce_sum应该理解为压缩求和，用于降维,按行求和 tf.negative()取反的即变负数

#预测，取得最近的邻居节点的下标
pred = tf.arg_min(distance,0)

#准确率
accuracy = 0
#变量初始化
init = tf.global_variables_initializer()

#启动图
with tf.Session() as sess:
    sess.run(init)
    #遍历测试数据集
    for i in range(len(X_test)):
        #获取最近的邻居节点
        nn_index = sess.run(pred, feed_dict = {train2_X:X_train,test2_X:X_test[i,:]})#feed_dict的作用是给使用placeholder创建出来的tensor赋值
        #获取最近的邻居节点的类别标签，并将其与该节点的真实类别标签进行比较
        print("测试数据", i ,"预测分类：",y_train[nn_index],"真实类别：",y_test[i])#返回最大值的索引
        #计算准确率
        if y_train[nn_index] == y_test[i]:
            accuracy += 1./len(X_test)
    print("分类准确率为：",accuracy)
```
